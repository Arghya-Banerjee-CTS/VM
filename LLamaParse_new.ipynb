```python
import os  # import required libraries
from llama_cloud_services import LlamaParse  # import required libraries
from llama_index.core import SimpleDirectoryReader  # import required libraries

os.environ['LLAMA_CLOUD_API_KEY'] = "<your-llamacloud-api-key>"  # set environment variable (API key, etc.)
llamaparse_api_key = os.getenv("LLAMA_CLOUD_API_KEY")  # parse the document using LlamaParse

parser = LlamaParse(  # initialize or use LlamaParse for document parsing
    api_key=llamaparse_api_key,  # parse the document using LlamaParse
    result_type="markdown",
    verbose=True
)

input_file = "/content/file-sample_150kB.pdf"
file_extractor = {".pdf": parser}  # parse the document using LlamaParse

documents = SimpleDirectoryReader(
    input_files=[input_file],
    file_extractor=file_extractor
).load_data()

output_dir = "parsed_output_simple"  # parse the document using LlamaParse
os.makedirs(output_dir, exist_ok=True)

output_file = os.path.join(
    output_dir, f"{os.path.splitext(os.path.basename(input_file))[0]}_parsed.md"  # parse the document using LlamaParse
)

markdown_content = "\n\n".join([doc.text for doc in documents])

with open(output_file, "w", encoding="utf-8") as f:
    f.write(markdown_content)  # save parsed output to file or folder

print(f"Parsed markdown saved to: {output_file}")  # display output or status message
```

```python
import os  # import required libraries
import nest_asyncio  # import required libraries
nest_asyncio.apply()

from IPython.display import Markdown, display  # import required libraries
from dotenv import load_dotenv  # import required libraries
load_dotenv()

# Import Azure OpenAI dependencies
from llama_index.llms.azure_openai import AzureOpenAI  # import required libraries
from llama_index.embeddings.huggingface import HuggingFaceEmbedding  # import required libraries
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings  # import required libraries
from llama_cloud_services import LlamaParse  # import required libraries


# Configure Azure OpenAI
llm = AzureOpenAI(
    engine=azure_deployment,
    api_key=azure_api_key,
    azure_endpoint=azure_endpoint,
    api_version=azure_api_version,
    temperature=1.0
)

# Configure Hugging Face local embedding model
embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Set global settings for llama-index
Settings.llm = llm
Settings.embed_model = embed_model

# Set up parser
parser = LlamaParse(  # initialize or use LlamaParse for document parsing
    api_key=llamaparse_api_key,  # parse the document using LlamaParse

    result_type="markdown"  # "markdown" and "text" are available
)

# Use SimpleDirectoryReader to parse our file
file_extractor = {".png": parser}  # parse the document using LlamaParse
input_file = '/content/images.png'
documents = SimpleDirectoryReader(input_files=[input_file], file_extractor=file_extractor).load_data()

# Save the parsed markdown locally
output_dir = 'parsed_output_image_11'  # parse the document using LlamaParse
os.makedirs(output_dir, exist_ok=True)  # Create output directory if it doesn't exist
output_file = os.path.join(output_dir, f"{os.path.splitext(os.path.basename(input_file))[0]}_parsed.md")  # parse the document using LlamaParse

# Extract markdown content from the parsed documents and save to file
markdown_content = "\n\n".join([doc.text for doc in documents])  # Combine all document texts
with open(output_file, 'w', encoding='utf-8') as f:
    f.write(markdown_content)  # save parsed output to file or folder

print(f"Parsed markdown saved to: {output_file}")  # display output or status message

# Create an index from the parsed markdown
index = VectorStoreIndex.from_documents(documents)

# Create a query engine for the index
query_engine = index.as_query_engine()

# Query the engine
query = "What is the total amount on the document?"
response = query_engine.query(query)

# Print the raw response
print("Query Response:", response)  # display output or status message
# display(Markdown(f"<b>{response}</b>"))
```
